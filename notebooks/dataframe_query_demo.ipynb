# Query Existing pandas and polars DataFrames

With the updated API, pass `dataframe=...` to `search(...)` to return matched rows with `similarity`.
Without `dataframe=...`, search keeps default numpy output behavior.


```python
import numpy as np
import pandas as pd
import polars as pl

from nndex import NNdex
```

```python
rng = np.random.default_rng(11)
rows, dims = 16, 8
matrix = rng.normal(size=(rows, dims)).astype(np.float32)
columns = [f"d{i}" for i in range(dims)]

pdf = pd.DataFrame(matrix, columns=columns)
pldf = pl.DataFrame({column: matrix[:, idx] for idx, column in enumerate(columns)})
index = NNdex(matrix, normalized=False, backend="cpu")
```

```python
# pandas: DataFrame output via dataframe=...
pandas_single = index.search(pdf.iloc[[0]], k=3, dataframe=pdf)
pandas_multi = index.search(pdf.iloc[[0, 1]], k=3, dataframe=pdf)
type(pandas_single), type(pandas_multi), len(pandas_multi), pandas_single
```

```python
# polars: DataFrame output via dataframe=...
polars_single = index.search(pldf[:1], k=3, dataframe=pldf)
polars_multi = index.search(pldf[:2], k=3, dataframe=pldf)
type(polars_single), type(polars_multi), len(polars_multi), polars_single
```

```python
# Default behavior without dataframe=... (numpy tuple output).
indices, scores = index.search(pdf.iloc[[0, 1]], k=3)
indices.shape, scores.shape
```
